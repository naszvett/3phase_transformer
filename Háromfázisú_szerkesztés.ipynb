```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lns
import math
```

Measurements

```python
A_B=225
A_C=228
B_C=227
A_a=158
C_a=238
A_c=157
C_c=159
```

Primer Triangle Vertexes

```python
# Primer
pA = (0,0)
pB = (A_C/2,(np.sqrt(3)*A_C)/2)
pC = (A_C,0)
```

Circles

```python
circles = []
radians = [A_a,A_c,C_a,C_c]
origoes = [pA,pA,pC,pC]
for i in range(4):
    circles.append(plt.Circle(origoes[i], radians[i], color='r', fill=False))
```

Secunder Triangle Vertexes

```python
def intersections(o1,r1,o2,r2):
    #origo distance
    d = np.sqrt((o1[0]-o2[0])**2+(o1[1]-o2[1])**2)
    if d==0:
        return None
    if d<abs(r1-r2):
        return None
    if d> r1+r2:
        return None
    a = (r1**2-r2**2+d**2)/(d*2)
    h = np.sqrt(r1**2-a**2)
    x = o1[0]+a*(o2[0]-o1[0])/d 
    y = o1[1]+a*(o2[1]-o1[1])/d
    
    #intersects
    ix1 = x+h*(o2[1]-o1[1])/d
    iy1 = y-h*(o2[0]-o1[0])/d
    ix2 = x-h*(o2[1]-o1[1])/d
    iy2 = y+h*(o2[0]-o1[0])/d
    
    return [(ix1,iy1),(ix2,iy2)]
```

```python
sa = intersections(pA,A_a,pC,C_a)[1]
sb = pB
sc = intersections(pA,A_c,pC,C_c)[1]
```

Drawing

```python
plt.gca().add_patch(circles[0])
plt.gca().add_patch(circles[1])
plt.gca().add_patch(circles[2])
plt.gca().add_patch(circles[3])
plt.gca().add_patch(plt.Polygon([pA,pB,pC], fill=False))
plt.gca().add_patch(plt.Polygon([sa,sb,sc], fill=False))

#Triangles
plt.gca().annotate("A",(pA[0]-15,pA[1]))
plt.gca().annotate("Bb",pB)
plt.gca().annotate("C",pC)
plt.gca().annotate("a",sa)
plt.gca().annotate("c",sc)

plt.gca().annotate("A-a és A-c kör",(pA[0]-A_a,pA[1]))
plt.gca().annotate("C-a kör",(pC[0]+C_a,pC[1]))
plt.gca().annotate("C-c kör",(pC[0]+C_c,pC[1]))
    
plt.plot()
```
